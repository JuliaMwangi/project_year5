## **CROP YIELD PREDICTION**

```python
import pandas as pd
import numpy as np
```

```python
#Load data
crop = pd.read_csv('D:/ACADEMICS/5TH YEAR/PROJECT/OUTPUTS/crop_yield.csv')
```

```python
crop.head()
```

### Data preprocessing and cleaning

```python
crop.shape
```

```python
crop.isnull().sum()
```

```python
crop.describe()
```

```python
crop1= crop.rename(columns={"Quantity.Ton..Annual":"Quantity"})
```

```python
crop1.columns
```

```python
crop1['Quantity'] = crop1['Quantity'].str.replace(',', '').astype(float)
```

```python
X = crop1.drop('Quantity',axis=1)
```

```python
X.head(1)
```

```python
#Convert te categorical column
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
X_transform = ohe.fit_transform(X[['SUBCOUNTY']])
X_transform.head()
```

```python
X_1 = pd.concat([X_transform,X],axis = 1).drop(['SUBCOUNTY'],axis = 1)
X_1.head()
```

```python
y= crop1['Quantity']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_1,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
#Model fitting
rf = RandomForestRegressor()
lr = LinearRegression()
```

```python
rf.fit(X_train,y_train)
```

```python
#Model prediction
y_pred = rf.predict(X_test)
```

```python
#Model evaluation
from sklearn.metrics import r2_score,mean_absolute_error as mae,mean_squared_error,mean_absolute_percentage_error
```

```python
acc = r2_score(y_test,y_pred)
acc
```

```python
mape1 = mean_absolute_percentage_error(y_test,y_pred)
mape1
```
